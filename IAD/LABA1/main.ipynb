```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Wholesale customers data.csv')
df.head()
```

```python
labels = df['Region']
data = df.drop(columns=['Region'])
data
```

```python
scaler = StandardScaler()
data = scaler.fit_transform(data)
data
```

```python
data_centered = data - data.mean()
cov_matrix = np.cov(data_centered, rowvar=False)
eig_values, eig_vectors = np.linalg.eig(cov_matrix)
idx = np.argsort(eig_values)[::-1]
eig_vectors = eig_vectors[:, idx]

projected_data_2d = data_centered.dot(eig_vectors[:, :2])
projected_data_3d = data_centered.dot(eig_vectors[:, :3])

# from sklearn.decomposition import PCA

# pca_2d = PCA(n_components=2)
# projected_data_2d = pca_2d.fit_transform(data)

# pca_3d = PCA(n_components=3)
# projected_data_3d = pca_3d.fit_transform(data)
```
